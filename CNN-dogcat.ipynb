```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D,Flatten,Dense
from tensorflow.keras import layers
import os
from google.colab import drive
```

```python
train_img = '/content/drive/MyDrive/train dog and cat'
test_img = '/content/drive/MyDrive/test dog and cat'
```

```python
print(os.listdir('/content/drive/MyDrive/train dog and cat'))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_data = ImageDataGenerator( rescale = 1/225)
test_data=ImageDataGenerator( rescale = 1/225)
train_imgnew = train_data.flow_from_directory(train_img,target_size=(28,28),class_mode='categorical')
test_imgnew = test_data.flow_from_directory(test_img,target_size=(28,28),class_mode='categorical')
```

Model
* The 1st layers has 16 filters , size 3*3
* The 2end layers has 4 filters , size 3*3


* layer Flatten
* Layers has 10 neurons


* Output have 2 labels
* 










```python
from keras.layers import BatchNormalization,Dropout
from keras.layers.attention.multi_head_attention import activation
model = keras.Sequential()
model.add(Conv2D(16,(3,3), activation='relu',input_shape=(28,28,3)))
model.add(Dropout(0.1,noise_shape=None, seed=None))
model.add(BatchNormalization(
    axis=1,scale='True'))
model.add(Conv2D(4 ,(3,3), activation='relu'))
model.add(Flatten())
model.add(Dense(10,activation='relu'))
model.add(Dense(2,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_imgnew,epochs=5)
```
